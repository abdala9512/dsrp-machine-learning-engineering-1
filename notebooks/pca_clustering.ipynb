# PCA + Clustering

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris_dataset = load_iris()

iris = pd.DataFrame(
    iris_dataset.data,
    columns=iris_dataset.feature_names
)
```

```python
pca = PCA(n_components=2)
pca_componets_v0 = pd.DataFrame(pca.fit_transform(iris), columns=["pc1", "pc2"])
```

```python
print(f"Varianza total epxlicada: {np.sum(pca.explained_variance_ratio_) * 100}")
```

```python
k_means = KMeans(n_clusters=2)
cluster_labels = k_means.fit_predict(pca_componets_v0)
pca_componets_v0["cluster" ] =  cluster_labels
```

```python
sns.scatterplot(x="pc1", y="pc2",hue="cluster", data=pca_componets_v0)
```
